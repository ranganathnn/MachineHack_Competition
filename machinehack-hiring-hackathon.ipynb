```python
train['InvoiceDate']=pd.to_datetime(train['InvoiceDate'])
test['InvoiceDate']=pd.to_datetime(test['InvoiceDate'])
```

```python
train['Quantity'] = train['Quantity'].abs()
test['Quantity'] = test['Quantity'].abs()
```

```python
train['Year'] = pd.DatetimeIndex(train['InvoiceDate']).year
train['Month'] = pd.DatetimeIndex(train['InvoiceDate']).month
train['Day'] = pd.DatetimeIndex(train['InvoiceDate']).day
#train['Hour'] = pd.DatetimeIndex(train['InvoiceDate']).hour
#train['minute'] = pd.DatetimeIndex(train['InvoiceDate']).minute

test['Year'] = pd.DatetimeIndex(test['InvoiceDate']).year
test['Month'] = pd.DatetimeIndex(test['InvoiceDate']).month
test['Day'] = pd.DatetimeIndex(test['InvoiceDate']).day
#test['Hour'] = pd.DatetimeIndex(test['InvoiceDate']).hour
#test['minute'] = pd.DatetimeIndex(test['InvoiceDate']).minute
```

```python
#train = train.drop(["InvoiceDate","Description", "InvoiceNo", "Quantity"], axis=1)
train = train.drop(["InvoiceDate", "StockCode", "InvoiceNo"], axis=1)
#train = train.drop("InvoiceDate", axis=1)

#test = test.drop(["InvoiceDate","Description", "InvoiceNo", "Quantity"], axis=1)
test = test.drop(["InvoiceDate", "StockCode", "InvoiceNo",], axis=1)
#test = test.drop("InvoiceDate", axis=1)
```

```python
X = train.drop("UnitPrice", axis=1)
```

```python
y = train["UnitPrice"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state = None)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model = GradientBoostingRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
model_score = model.score(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
import numpy as np
```

```python
print("Root Mean squared error: %.2f"% np.sqrt(mean_squared_error(y_test, y_predicted)))
```
